```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
dataset_1 = np.loadtxt("bz1-s_blyp_tot_ener.txt")
```

```python
plt.plot(dataset_1[:,0],
         dataset_1[:,1]-np.min(dataset_1[:,1]),
         "--x",
         markersize=5,
         label="HF/cc-pVDZ")

plt.plot(dataset_1[:,0],
         dataset_1[:,1]-np.min(dataset_1[:,1]),
         "--s",
         markersize=5,
         label="BLYP/cc-pVDZ")

plt.plot(dataset_1[:,0],
         dataset_1[:,1]-np.min(dataset_1[:,1]),
         "--o",
         markersize=5,
         label="B3LYP/cc-pVDZ")

plt.legend()
plt.xlabel("< x-axis label -- remember units! >")
plt.ylabel("< y-axis label -- remember units! >")
plt.ylim([0,6])
plt.xticks(np.arange(0,390,30))


plt.savefig("plot-1.png")

```
